```python
📅 5. Temporal or Launch-based Insights

Are newer cars (Latest_Launch) selling better?
```

```python
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'], format='%m/%d/%Y')
min_date = df['Latest_Launch'].min()
df['Days_Since_Min_Launch'] = (df['Latest_Launch'] - min_date).dt.days
correlation = df['Days_Since_Min_Launch'].corr(df['Sales_in_thousands'])
print(f"Correlation between Car Launch Date (Newer) and sales: {correlation:.4f}")
```

```python
#How have average prices or sales changed over time (if Latest_Launch is convertible to year)?
```

```python
df['Launch_Year'] = df['Latest_Launch'].dt.year
yearly_trend = df.groupby('Launch_Year')[['Price_in_thousands', 'Sales_in_thousands']].mean()
yearly_trend = yearly_trend.sort_index()
print("Average Price and Sales Trend by Launch Year:")
print(yearly_trend.to_string(float_format='%.2f'))
```

```python
plot_cols = ['Price_in_thousands', 'Sales_in_thousands', 'Horsepower', 'Fuel_efficiency']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(plot_cols):
    sns.histplot(
        data=df,
        x=col,
        kde=True,
        ax=axes[i],
        bins=20,
        color='skyblue' if i % 2 == 0 else 'salmon'
    )
    axes[i].set_title(f'Distribution of {col}', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Frequency / Density', fontsize=12)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('distribution_plots.png')
    
```

```python
#Correlation heatmap for numeric features
```

```python
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 10))
mask = np.triu(correlation_matrix)

sns.heatmap(
    correlation_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=.5,
    cbar_kws={"shrink": .8}
)

plt.title('Correlation Heatmap of Numerical Features', fontsize=16)

plt.savefig('correlation_heatmap.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate total sales by manufacturer
sales_by_manufacturer = df.groupby('manufacturer')['sales'].sum().sort_values(ascending=False).reset_index()

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=sales_by_manufacturer, x='manufacturer', y='sales', palette='viridis')
plt.xticks(rotation=45)
plt.title('Total Sales by Manufacturer', fontsize=16)
plt.ylabel('Total Sales')
plt.xlabel('Manufacturer')
plt.show()
```
